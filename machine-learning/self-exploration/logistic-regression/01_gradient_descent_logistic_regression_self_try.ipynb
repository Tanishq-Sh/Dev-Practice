```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    
    g = 1 / (1 + np.exp(-z))
    
    return g

def compute_gradient(X, y, w, b):
    
    m, n = X.shape
    
    dj_dw = np.zeros(n)
    dj_db = 0.
    
    for i in range(m):
        
        f_wb_i = sigmoid(np.dot(X[i], w) + b)
        err = f_wb_i - y[i]
        
        for j in range(n):
            
            dj_dw[j] += err * X[i, j]
            
        dj_db += err
        
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw, dj_db
    
    

def gradient_descent(X, y, w, b, alpha, n_iters):
    
    w_copy = w
    b_copy = b
    
    for i in range(n_iters):
        
        dj_dw, dj_db = compute_gradient(X, y, w_copy, b_copy)
        w_copy = w_copy - alpha * dj_dw
        b_copy = b_copy - alpha * dj_db
        
    return w_copy, b_copy
```

```python
# Training set 1
X_train = np.array([[0.5, 1.5],[1, 1],[1.5, 0.5],[3, 0.5],[2, 2],[1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# Training set 2
# X_train = np.array([0., 1, 2, 3, 4, 5]).reshape(-1,1)
# y_train = np.array([0, 0, 0, 1, 1, 1])
```

## Initialize weights and run gradient descent for logistic regression

```python
# training set 1
w_init = [1, 1]
b_init = 5
alpha = 0.001
n_iters = 100000

# training set 2
# w_init  = np.zeros_like(X_train[0])
# b_init  = 0.
# alpha = 0.1
# n_iters = 10000

w_final, b_final = gradient_descent(X_train, y_train, w_init, b_init, alpha, n_iters)
print(w_final, b_final)

# training set 1
x0 = -b_final/w_final[0]
x1 = -b_final/w_final[1]

# training set 2
# x0 = -b_final/w_final[0]
# x1 = b_final

print(x0, x1)
```
